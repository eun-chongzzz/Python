주피터 노트북 사용법

셀을 클릭하면 녹색으로 선택 영역이 지정됨.<br>
녹색으로 선택된 셀에는 키보드 입력시 코드/입력문이 직접 입력됨<br>
녹생으로 선택된 셀에 esc를 입력하거나 혹은 바깥쪽 셀만 선택시 파란색으로 선택됨.<br>

파란색으로 선택된 셀에<br>
b 입력 => 선택된 셀 아래에 입력용 셀 추가<br>
a 입력 => 선택된 셀 위에 입력용 셀 추가<br>
dd(입력) => 선택된 셀 삭제

m => 선택된 셀을 코드 입력용이 아닌 필기용으로 전환<br>
y => 선택된 셀을 필기용에서 코드입력용으로 전환<br>

필기용 셀은 shift + enter를 누르면 필기가 완료된 것으로 간주합니다.

# 주피터노트북 필기
## #키를 이용하면 글씨크기 조절이
### 가능합니다.
#### # 갯수가 많아지면 많아질수록
##### 글씨크기는 작아집니다.
###### #은 6개까지 쓸 수 있습니다.
<hr>

-나열해서 필기하고 싶다면
- -한칸 뒤에 필기를 합니다.
- 이렇게 나열이 됩니다.
    - 띄어쓰기 네 칸 후에 -를 붙이면
    - 도트 하위 항목으로 간주됩니다.
        - 하위 항목에서 다시 띄어쓰기를 네 번 더 하면
        - 하위의 하위 항목이 됩니다.
